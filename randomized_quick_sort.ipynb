### Randomized quick sort implementation

```python
import random
qs_list = random.sample(range(1, 1000), 100)
```

```python
def choose_pivot(l,r):
    p = random.randint(l,r)
    return p
```

```python
def partition(A,l,r): #A:list to process, l:left boundry, r:right boundry, p:pivot
    n_elements = r-l
    print("r,l,n elements ", str(r),str(l),str(n_elements))
    i=0
    if n_elements<=1:
        return
    if n_elements>1:
        p = choose_pivot(l,r-1)
        print("p,l,r",p,l,r)
        pivot = A[p]
        print('pivot='+str(pivot))
        print("list before swap"+str(A))
        A[l],A[p] = A[p],A[l] #Swap pivot to the first element
        print("list after swap"+str(A))

        i = l+1
        for j in range(l+1,r):#scan through all elements except for the pivot
            if A[j]<pivot:
                A[i],A[j] = A[j],A[i]
                i +=1
        A[l],A[i-1] = A[i-1],A[l]
        print("after pivot by " + str(pivot) + " " +str(A))
        
        pivot_sorted = i-1
        
        partition(A,l,pivot_sorted)  
        partition(A,pivot_sorted,r)
    return i-1
```
